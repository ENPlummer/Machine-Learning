```python
import pandas as pd
import numpy as np
```

```python
#Read in the data into a dataframe using pd.read_table.
data = pd.read_table("house-votes-84.data", sep=",", header=None)
data
```

```python
#Replace the questions marks, "?" in the data with NaN by using .replace() method.
vote_data = data.replace("?", np.NaN)
vote_data
```

```python
vote_names = vote_data.rename(columns={0:"Party", 1:"Disabled Infants", 2:"Water Project", 3:"Budget Resolution", 4:"Physician Fee Freeze", 5:"El Salvador Aid",6:"Religious Groups in Schools", 7:"Anti Satellite Test Ban",
                                       8:"Aid to Nicaraguan Contras", 9:"MX Missile", 10: "Immigration", 11:"Synfuels Corporation Cutback", 12:"Education Spending", 13:"Superfund Right to Sue",
                                      14:"Crime", 15:"Duty Free Exports",16:"Export Administration Act South Africa"})
vote_names
```
